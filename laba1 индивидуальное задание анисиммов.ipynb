```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
url = "https://raw.githubusercontent.com/adel1726/laba1/refs/heads/main/machine.data"
data = pd.read_csv(url, header=None)

# Предполагаем, что данные имеют 10 столбцов
vendor_name = data[0]
model_name = data[1]
myct = data[2]
mmin = data[3]
mmax = data[4]
cach = data[5]
chmin = data[6]
chmax = data[7]
prp = data[8]
erp = data[9]

# Список параметров для построения графиков
parameters = [myct, mmin, mmax, cach, chmin, chmax, prp, erp]
labels = ['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX', 'PRP', 'ERP']

# Создание графиков
fig, axs = plt.subplots(len(parameters), 2, figsize=(12, 20))

# Графики для VendorName
for i, param in enumerate(parameters):
    axs[i, 0].scatter(vendor_name[:50], param[:50], color='r', label='Vendor 1')
    axs[i, 0].scatter(vendor_name[50:100], param[50:100], color='g', label='Vendor 2')
    axs[i, 0].scatter(vendor_name[100:150], param[100:150], color='b', label='Vendor 3')
    axs[i, 0].set_xlabel('Vendor Name')
    axs[i, 0].set_ylabel(labels[i])
    axs[i, 0].legend()
    axs[i, 0].set_xticks(range(len(vendor_name.unique())))
    axs[i, 0].set_xticklabels(vendor_name.unique(), rotation=45)

# Графики для ModelName
for i, param in enumerate(parameters):
    axs[i, 1].scatter(model_name[:50], param[:50], color='r', label='Vendor 1')
    axs[i, 1].scatter(model_name[50:100], param[50:100], color='g', label='Vendor 2')
    axs[i, 1].scatter(model_name[100:150], param[100:150], color='b', label='Vendor 3')
    axs[i, 1].set_xlabel('Model Name')
    axs[i, 1].set_ylabel(labels[i])
    axs[i, 1].legend()
    # Устанавливаем фиксированные метки для оси X
    axs[i, 1].set_xticks(range(len(model_name.unique())))
    axs[i, 1].set_xticklabels(model_name.unique(), rotation=45)

plt.tight_layout()
plt.show()
```
