```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Создаем пример данных
train_data = {
    'Platform': ['Wii', 'PS3', 'DS', 'X360', 'PC'],
    'Year': [2006, 2007, 2004, 2012, 2011],
    'Global_Sales': [82.74, 28.5, 30.26, 25.3, 12.1]
}
test_data = {
    'Platform': ['GBA', 'PS4'],  # Новые платформы, которых нет в обучающих данных
    'Year': [2005, 2013],
    'Global_Sales': [33.2, 18.7]
}

df_train = pd.DataFrame(train_data)
df_test = pd.DataFrame(test_data)

# 2. Разделяем данные
X_train = df_train[['Platform', 'Year']]
y_train = df_train['Global_Sales']
X_test = df_test[['Platform', 'Year']]
y_test = df_test['Global_Sales']

# 3. Создаем пайплайн с обработкой новых категорий
preprocessor = ColumnTransformer(
    transformers=[
        ('platform', OneHotEncoder(handle_unknown='ignore'), ['Platform']),  # Важно: handle_unknown='ignore'
        ('year', SimpleImputer(strategy='median'), ['Year'])
    ])

model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# 4. Обучаем модель
model.fit(X_train, y_train)

# 5. Делаем предсказания (теперь сработает, даже с новыми категориями)
try:
    y_pred = model.predict(X_test)
    print("Предсказания успешно выполнены:")
    print(y_pred)
except Exception as e:
    print(f"Ошибка при предсказании: {e}")

# 6. Дополнительная информация
print("\nСтатус обработки новых категорий:")
print(f"Обучающие категории: {model.named_steps['preprocessor'].transformers_[0][1].categories_}")
print(f"Встреченные новые категории: ['GBA', 'PS4'] были проигнорированы")
```
